## Valores Nulos

```python
# verificando se existem registros (linhas) do dataframe com valor vazio (nulo)
df.isnull().sum()
```

```python
# podemos remover a linha com valores nulos com o comando abaixo
# df.dropna(inplace=True)

# ou conhecemos melhor sobre esse registro
df.query('`MARCA DO VEÍCULO`.isnull()', engine='python')
```

```python
# como sabemos, a SPORTAGE é da Kia então:
# somente um registro é nulo então vou substituir o valor nulo por Kia
# IMPORTANTE, cada caso é um caso.

df.fillna('Kia', inplace=True)

# verificar senão há mais nenhum valor como nulo
df.isna().sum()
```

## Valores únicos

```python
# primeiro, vamos contar quantos registros temos no nosso dataframe
colunas = ['MARCA DO VEÍCULO', 'ANO DO VEÍCULO']
df[colunas].shape
```

```python
# agora vamos fazer a mesma coisa mas não quero valores duplicados nas colunas
colunas = ['MARCA DO VEÍCULO', 'ANO DO VEÍCULO']
df[colunas].drop_duplicates().shape
```

## Valores baseado em um padrão

```python
# selecionar todos os veículos que contém a string "lass" no nome
colunas = ['VEÍCULO', 'ANO DO VEÍCULO']
df.query('`VEÍCULO`.str.contains("lass")').head()[colunas]
```

## Concatenação

```python
# liste os primeiros 5 veículos, exibindo somente as colunas MARCA DO VEÍCULO e ANO DO VEÍCULO
# e criando uma NOVA COLUNA com os valores das duas colunas concatenados
colunas = ['MARCA DO VEÍCULO', 'ANO DO VEÍCULO']

# df1 recebe uma cópia das 5 primeiras linhas
df1 = df[colunas].head().copy()
display(df1)
print("\n")

# df2 faz a concatenação dos valores das 2 colunas
df2 = df1[colunas].apply(lambda x: str(x['MARCA DO VEÍCULO']) +' '+ str(x['ANO DO VEÍCULO']), axis=1)
display(df2)
print("\n")

# criando a NOVA COLUNA no dataframe df1 com os dados do df2
df1['NOVA COLUNA'] = df2
display(df1)
```

## Valores calculados

```python
# faça uma cópia dos primeiros 5 veículos, exibindo as colunas abaixo em
# um dataframe chamado df1
colunas = ['VALOR BRUTO', 'VALOR DO DESCONTO', 'VALOR LIQUIDO']
df1 = df[colunas].head().copy()
display(df1)

# crie um df2 que calcule a subtração das duas primeiras colunas
df2 = df1.apply(lambda x: x['VALOR BRUTO'] - x['VALOR DO DESCONTO'], axis=1)

# crie uma NOVA COLUNA no df1 com o resultado do df2
df1['NOVA COLUNA'] = df2

display(df1)
```

```python
# calculando a soma da NOVA COLUNA
df1['NOVA COLUNA'].sum()
```

```python
# calculando o menor valor da NOVA COLUNA
df1['NOVA COLUNA'].min()
```

```python
# calculando o maior valor da NOVA COLUNA
df1['NOVA COLUNA'].max()
```

```python
# calculando a média da NOVA COLUNA
df1['NOVA COLUNA'].mean()
```

```python
# calculando a mediana da NOVA COLUNA
df1['NOVA COLUNA'].median()
```

```python
# fazendo tudo isso de uma única só vez
df1['NOVA COLUNA'].describe()
```

## Valores agrupados

```python
# criando um dataframe com 30 registros (linhas) aleatórios do df
colunas = ['MARCA DO VEÍCULO', 'VALOR BRUTO']
df1 = df[colunas].sample(30)

# somar os valores agrupando-os por marca
df1.groupby('MARCA DO VEÍCULO').sum()
```

```python
# vamos repetir o agrupamento mas iremos ordenar pelo valor bruto
# da maior para a menor soma
df1.groupby('MARCA DO VEÍCULO').sum().sort_values(by='VALOR BRUTO', ascending=False)
```

# **Filtros**

```python
# filtrando dados operadores
#  > (maior)    < (menor)     >= (maior ou igual)     <= (menor ou igual)
colunas = ['DATA DA VENDA', 'NOME DO VENDEDOR', 'VALOR LIQUIDO']
df[df['VALOR LIQUIDO'] > 13800][colunas]
```

```python
# OU usando a função query
df.query('`VALOR LIQUIDO` > 13800')[colunas]
```

```python
# RELEMBRANDO AS COLUNAS DO DATAFRAME df
df.sample(1)
```

```python
# filtrando com operador & (AND)
df.query('`VALOR DO DESCONTO` >= 900 & `MARCA DO VEÍCULO` == "FIAT"' )
# explicando: selecionar todos os dados que tem o valor do desconto
# igual ou maior que 900 e a marca é FIAT
```

```python
# filtrando com operador | (OR)
df.query('`VALOR DO DESCONTO` >= 900 & (`MARCA DO VEÍCULO` == "FIAT" | `MARCA DO VEÍCULO` == "MINI")' )
# explicando: selecionar todos os dados que tem o valor do desconto
# igual ou maior que 900 e a marca é (FIAT OU MINI)
```

# **Ordenação**

```python
# ordenando a seleção de dados por ordem alfabética dos veículos.
# o caracter \ é somente para informar ao interpretador python que o comando continua na próxima linha
# usei a função .head(10) para exibir as 10 primeiras linhas

colunas = ['VEÍCULO', 'VALOR LIQUIDO'] # exibir somente as colunas VEÍCULO e VALOR LÍQUIDO

df.sort_values(by='VEÍCULO', ascending=True) \
.head(10)[colunas]
```

```python
# a mesma seleção acima mas ordenando por nome e por valor do veículo, porém
# o valor será ordenado do maior para o menos, ou seja, decrescente

colunas = ['VEÍCULO', 'VALOR LIQUIDO'] # exibir somente as colunas VEÍCULO e VALOR LÍQUIDO

df.sort_values(by='VALOR LIQUIDO', ascending=False) \
.sort_values(by='VEÍCULO', ascending=True) \
.head(10)[colunas]
```
